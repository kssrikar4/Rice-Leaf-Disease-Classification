```python
import zipfile
import os
from glob import glob
from collections import Counter
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Mount / unzip – adjust paths
zips = [
    "anshulm257-rice-disease-dataset.zip",
    "vbookshelf-rice-leaf-diseases.zip",
    "nirmalsankalana-rice-leaf-disease-image.zip"
]

for z in zips:
    with zipfile.ZipFile(z, 'r') as zip_ref:
        zip_ref.extractall("./rice_data")
```

```python
def get_class_counts(root_path):
    class_counts = Counter()
    for class_dir in glob(os.path.join(root_path, "*")):
        if os.path.isdir(class_dir):
            class_name = os.path.basename(class_dir)
            num = len(glob(os.path.join(class_dir, "*.jpg")) + 
                      glob(os.path.join(class_dir, "*.png")) +
                      glob(os.path.join(class_dir, "*.JPG")))
            class_counts[class_name] = num
    return class_counts

# Apply to each extracted folder
for dataset_folder in glob("./rice_data/*"):
    if os.path.isdir(dataset_folder):
        print(f"\n📁 {dataset_folder}")
        counts = get_class_counts(dataset_folder)
        for cls, c in counts.items():
            print(f"   {cls}: {c} images")
```

```python
sample_images = glob("./rice_data/**/*.jpg", recursive=True)[:100]
sizes = []
for img_path in sample_images:
    img = cv2.imread(img_path)
    if img is not None:
        sizes.append(img.shape[:2])  # (height, width)

unique_sizes = set(sizes)
print(f"Unique image dimensions: {unique_sizes}")
```

```python
import pandas as pd

all_classes = {}
for dataset_folder in glob("./rice_data/*"):
    if os.path.isdir(dataset_folder):
        folders = [d for d in glob(os.path.join(dataset_folder, "*")) if os.path.isdir(d)]
        class_names = [os.path.basename(f) for f in folders]
        all_classes[dataset_folder] = class_names

df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in all_classes.items()]))
print(df)
```
